```python
# import the required modules
import matplotlib.pyplot as plt
int datetime as dt 
```

```python
# Starting point
start_date = dt.date(2025, 11, 3)
current_problems = 163

# Target and estimates
targets = [200, 300, 400, 500]
problems_left = [t - current_problems for t in targets]
```

```python
# Rates
rate_fast = 1.5     # problems/day
rate_slow = 1.2     # problems/day

# Calculate estimated completion dates
dates_fast = [start_date + dt.timedelta(days=p/rate_fast) for p in problems_left]
dates_slow = [start_date + dt.timedelta(days=p/rate_slow) for p in problems_left]
```

```python
# Plot 
plt.figure(figsize=(10,5))
plt.plot(dates_fast, targets, marker='o', label="At 1.5 problems/day (current pace)", color = "green")
plt.plot(dates_fast, targets, marker='o', label="At 1.2 problems/day (slower pace)", color = "orange")
plt.axvline(start_date,color='gray', linestyle='--',alpha=0.6)
```

```python
# Label and styling
plt.title("Leetcode Progress Projection (from 163 problems on Nov 3, 2025)", fontsize=13)
plt.xlabel("Date")
plt.ylabel("Total Problems Solved")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

plt.show()
```

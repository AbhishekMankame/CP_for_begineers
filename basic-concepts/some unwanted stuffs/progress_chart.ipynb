```python
# Starting point
start_date = dt.date(2025, 11, 3)
current_problems = 163

# Target and estimates
targets = [200, 300, 400, 500]
problems_left = [t - current_problems for t in targets]
```

```python
# Rates
rate_fast = 1.5     # problems/day
rate_slow = 1.2     # problems/day

# Calculate estimated completion dates
dates_fast = [start_date + dt.timedelta(days=p/rate_fast) for p in problems_left]
dates_slow = [start_date + dt.timedelta(days=p/rate_slow) for p in problems_left]
```
